```python
import pandas as pd

nme = ['Google','Runoob','Taobao', 'Wiki']
st = ["www.google.con","www.runoob.com","www.taobao.com","www.wikipedia.org"]
ag = [90,40,80,98]

dict = {'name':nme,'site':st,'age':ag}
df = pd.DataFrame(dict)
(df)
df.to_csv('site.csv')

```

```python
# 示例数据
data2 = {'Name': ['Alice', 'Bob', 'Charlie', None],
        'Age': [25, 30, None, 35],
        'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']}

df2 = pd.DataFrame(data2)

# 填充缺失的 "Age" 为均值
df2['Age'].fillna(df2['Age'].mean(), inplace=True)

print(df2)
```
